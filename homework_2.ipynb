```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from ast import literal_eval
```

```python
data = pd.read_csv('data.csv')
```

```python
def cat(df, spec):
    table = df.pivot_table(index=spec, columns='Type', aggfunc='size', fill_value=0)

    cat = table.index
    status = table.columns
    n_cat = len(cat)
    width = 1
    space_between = 3
    x = np.arange(0, n_cat * space_between, space_between)

    fig, ax = plt.subplots(figsize=(60, 20))
    for i, stat in enumerate(status):
        ax.bar(x + i * width, table[stat], width, label=stat, color=['red' if stat == 1 else 'green'])

    ax.set_title('Вредоносность сайта', pad=20, fontsize=14)
    ax.set_xlabel(spec, fontsize=12)
    ax.set_ylabel('Количество сайтов', fontsize=12)
    ax.set_xticks(x + width/2)
    ax.set_xticklabels(cat, fontsize=12, rotation=90)
    ax.legend(title='Статус', fontsize=11)
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    
    for i, stat in enumerate(status):
        for j, value in enumerate(table[stat]):
            ax.text(x[j] + i*width, value + 0.5, str(value), 
                    ha='center', va='bottom', fontsize=11)
    plt.show()
```

```python
cat_columns = ['CHARSET', 'SERVER', 'WHOIS_COUNTRY', 'WHOIS_STATEPRO', 'WHOIS_REGDATE']
for i in cat_columns:
    cat(data, i)
```

```python
def number(bad, good, column):
    plt.figure(figsize=(30, 6))
    plt.hist(bad[column], bins=100, edgecolor='black', alpha=1, color='red', label="Вредоносный")
    plt.hist(good[column], bins=100, edgecolor='black', alpha=0.5, color='green', label="Защищённый")
    plt.xlabel(f"Характеристика {column} сайта")
    plt.ylabel("Количество")
    plt.title(f"Сравнение распределения {column} вредоносных и защищенных сайтов")
    plt.legend()
    plt.show()
```

```python
data['CONTENT_LENGTH'] = data.CONTENT_LENGTH.fillna(data.CONTENT_LENGTH.median())
data['DNS_QUERY_TIMES'] = data.DNS_QUERY_TIMES.fillna(data.DNS_QUERY_TIMES.median())
```

```python
bad_sites = data[data['Type'] == 1]
good_sites = data[data['Type'] == 0]
num_column = data.columns

for column in num_column[1:-1]:
    if column not in ['URL', 'CHARSET', 'SERVER', 'WHOIS_COUNTRY', 'WHOIS_STATEPRO', 'WHOIS_REGDATE', 'WHOIS_UPDATED_DATE']:
        number(bad_sites, good_sites, column)
```
